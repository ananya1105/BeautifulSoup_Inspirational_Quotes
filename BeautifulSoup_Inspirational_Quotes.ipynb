```python
#Python program to scrape website  
#and save quotes from website 
import requests 
from bs4 import BeautifulSoup 
import csv 
  
URL = 'https://motivationping.com/quotes/'
r = requests.get(URL) 
  
soup = BeautifulSoup(r.content, 'html5lib') 
print(soup)
quotes=[]  # a list to store quotes 
  
table = soup.findAll('strong') 
print(table)
for row in table: 
    quotes.append(row.text)
MyFile=open('C:/Users/hp/Desktop/output1.txt','w')    
filename = 'inspirational_quotes.txt'
#with open(filename, "wt") as output:
    
#    for q in quotes:
#        output.write(q)

for q in quotes:
     MyFile.write(q)
     MyFile.write('\n')
MyFile.close()
```
